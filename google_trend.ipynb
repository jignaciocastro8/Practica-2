```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
plt.style.use("ggplot")
```

```python
online_compras = pd.read_csv("data/multiTimeline.csv")[1:]["Categoría: Todas las categorías"]
values = online_compras.values
fechas = online_compras.index
df = pd.DataFrame()
df["fechas"] = fechas
df["values"] = values
#df["values"] = df["values"].rolling(window = 5).mean()
df.head()
```

```python
def str_date(row):
    l = row.split("-")
    return datetime.datetime(int(l[0]), int(l[1]), int(l[2])).date()
df["fechas"] = df["fechas"].apply(str_date)
df = df.set_index(df["fechas"])
df = df.drop(["fechas"], axis = 1)
df = df.reindex(pd.date_range("2016-04-17", "2021-04-11"), method='ffill')
#df["fechas"] = df.index
df = df[df.index <= "2020-12-30"]
df = df["2019-01-01" <= df.index]
df["values"] = df["values"].rolling(window=14).mean()
```

```python
plt.figure(figsize=(10,8))
plt.plot(df["values"])
plt.xticks(rotation=45)
plt.title("Google trends: compra online")
plt.ylabel("Popularidad")
plt.savefig("informe/imgs/google_trend_compra_online.png", dpi=600)
```
